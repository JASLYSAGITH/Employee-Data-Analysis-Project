*  Identify the predominant age group among employees.

```python
pre_age=DF["Age"].value_counts()
pre_age
```

```python
bins = [18, 24, 34, 44, 54, 64]
labels = ['18-24', '25-34', '35-44', '45-54', '55-65']
DF['age_group'] = pd.cut(DF['Age'], bins=bins, labels=labels, right=False)
DF_pre_age=DF['age_group'].value_counts()

DF_pre_age    
```

✍️ After above operations we concluded that the predominant age group is 25-34


Insights:👁️👁️👁️
- The 25-34 age group clearly stands out as the predominant group, with 309 employees.
- This indicate that the workforce is largely comprised of individuals in the early to mid-career stage.
- This could indicate that the company is attracting younger talent, likely early-career professionals who are in their prime working age,
   and have a high level of energy, adaptability, and potential for growth within the company.
- The 18-24 age group has 107 employees, which is a relatively strong number, showing that the company is likely attracting new graduates 
   or young professionals who are just starting their careers.
- No employees in the 45-54 and 55-65 age groups 
- The absence of employees in these age groups could also be a sign of a younger, dynamic workforce.but it shows the absence of experienced employees.

* Investigate if there's any correlation between age and salary, and represent it visually

```python
correlation_matrix = DF[['Age', 'Salary']].corr()
correlation_matrix
```

Insights:👁️👁️👁️

- The diagonal values (1.000 for both "Age" and "Salary") represent the correlation of each variable with itself. This is always 1.00 because a variable is always perfectly correlated with itself.
- The correlation between Age and Salary is 0.211.
- This is a positive but weak correlation, suggesting that there is a slight tendency for salary to increase as age increases.
- However, the correlation is low, so this relationship is not strong and might not be statistically significant on its own.


```python
plt.figure(figsize=(6,6))
sns.heatmap(correlation_matrix, annot=True, cmap='twilight', center=0)
plt.title('Correlation Heatmap Between Age and Salary', color='blue',fontweight='bold')
plt.show()
```

✍️ interpretation :
The correlation coefficient (0.211) suggests that age and salary are weakly positively correlated. 
This could mean that, in the dataset, older individuals tend to have slightly higher salaries, but the relationship is not strong

✍️ Based on this weak correlation, age might not be a strong predictor of salary, and other factors might play a more significant 
role in determining salary.
